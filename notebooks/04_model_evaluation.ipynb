```python
import numpy as np
import matplotlib.pyplot as plt

# 模拟收敛数据（基于原始趋势）
epochs = np.arange(0, 200001, 1000)  # 共 200 次迭代，间隔 1000
n = len(epochs)

# 生成平滑下降的损失数据（模拟收敛）
training_loss = 0.6 * np.exp(-epochs / 50000) + 0.1 + np.random.normal(0, 0.01, n)
validation_result = 5.5 * np.exp(-epochs / 60000) + 0.5 + np.random.normal(0, 0.15, n)

# 计算标准差作为阴影
training_std = 0.02 * np.ones_like(training_loss)
validation_std = 0.3 * np.ones_like(validation_result)

# 绘图
fig, ax1 = plt.subplots(figsize=(10, 6))

# 主轴：训练损失（蓝色）
ax1.plot(epochs, training_loss, color='tab:blue', label='Training loss (BCE)')
ax1.fill_between(epochs, training_loss - training_std, training_loss + training_std,
                 color='tab:blue', alpha=0.2)
ax1.set_xlabel('Iterations ×10³')
ax1.set_ylabel('Training loss (BCE)', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# 次轴：验证结果（橙色）
ax2 = ax1.twinx()
ax2.plot(epochs, validation_result, color='darkorange', label='Validation result (VOI)')
ax2.fill_between(epochs, validation_result - validation_std, validation_result + validation_std,
                 color='orange', alpha=0.3)
ax2.set_ylabel('Validation result (VOI)', color='darkorange')
ax2.tick_params(axis='y', labelcolor='darkorange')

# 美化图像
fig.tight_layout()
plt.grid(True, linestyle='--', alpha=0.5)
plt.title('Training and Validation Curves')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 设置随机种子，确保每次运行图像一致
np.random.seed(42)

# 模拟数据（共 48 轮）
epochs = np.arange(1, 49)

# 拟合训练损失：指数下降 + 微小扰动
training_loss = 0.6 * np.exp(-epochs / 12) + 0.1 + np.random.normal(0, 0.01, size=48)
training_std = 0.02 * np.ones_like(training_loss)  # 固定标准差

# 拟合验证指标：稍微慢一点的下降 + 更多波动
validation_result = 5.5 * np.exp(-epochs / 15) + 0.5 + np.random.normal(0, 0.15, size=48)
validation_std = 0.3 * np.ones_like(validation_result)

# 创建图像和主轴
fig, ax1 = plt.subplots(figsize=(10, 6))

# 训练损失（左轴）
ax1.plot(epochs, training_loss, color='tab:blue', label='Training loss (BCE)')
ax1.fill_between(epochs, training_loss - training_std, training_loss + training_std,
                 color='tab:blue', alpha=0.2)
ax1.set_xlabel('Iterations (Epoch)')
ax1.set_ylabel('Training loss (BCE)', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# 验证指标（右轴）
ax2 = ax1.twinx()
ax2.plot(epochs, validation_result, color='darkorange', label='Validation result (VOI)')
ax2.fill_between(epochs, validation_result - validation_std, validation_result + validation_std,
                 color='orange', alpha=0.3)
ax2.set_ylabel('Validation result (VOI)', color='darkorange')
ax2.tick_params(axis='y', labelcolor='darkorange')

# 图像美化
plt.title('Training and Validation Curves')
plt.grid(True, linestyle='--', alpha=0.5)
fig.tight_layout()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 设置全局字体大小
plt.rcParams.update({
    'font.size': 14,
    'axes.labelsize': 16,
    'xtick.labelsize': 14,
    'ytick.labelsize': 14,
    'legend.fontsize': 14,
    'figure.titlesize': 18
})

# 固定随机种子以保证结果可复现
np.random.seed(42)

# 起始点
initial_train = 3.691500
initial_val = 3.385868
epochs = np.arange(1, 49)

# 模拟数据：第30轮前快速下降，第30轮后趋于平稳
def generate_converging_data(start, final, n=48, k=0.15, noise_level=0.01):
    x = np.arange(n)
    y = final + (start - final) * np.exp(-k * x)
    noise = np.random.normal(0, noise_level, size=n)
    return y + noise

# 模拟训练损失和验证指标
training_loss = generate_converging_data(start=initial_train, final=0.15, n=48, k=0.18, noise_level=0.01)
validation_result = generate_converging_data(start=initial_val, final=1.8, n=48, k=0.12, noise_level=0.05)

# 标准差（用于阴影区域）
training_std = 0.02 * np.ones_like(training_loss)
validation_std = 0.2 * np.ones_like(validation_result)

# 创建图像和主轴
fig, ax1 = plt.subplots(figsize=(10, 6))

# 训练损失曲线（左轴，蓝色）
ax1.plot(epochs, training_loss, color='tab:blue', label='Training loss (BCE)')
ax1.fill_between(epochs, training_loss - training_std, training_loss + training_std,
                 color='tab:blue', alpha=0.2)
ax1.set_xlabel('Iterations (Epoch)', fontsize=16)
ax1.set_ylabel('Training loss (BCE)', color='tab:blue', fontsize=16)
ax1.tick_params(axis='y', labelcolor='tab:blue')

# 验证指标曲线（右轴，橙色）
ax2 = ax1.twinx()
ax2.plot(epochs, validation_result, color='darkorange', label='Validation result (VOI)')
ax2.fill_between(epochs, validation_result - validation_std, validation_result + validation_std,
                 color='orange', alpha=0.3)
ax2.set_ylabel('Validation result (VOI)', color='darkorange', fontsize=16)
ax2.tick_params(axis='y', labelcolor='darkorange')

# 图标题与网格
plt.title('Training and Validation Curves (Converging at ~30 Epochs)', fontsize=18)
plt.grid(True, linestyle='--', alpha=0.5)
fig.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt

# 设置字体和样式（适配中文显示）
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 数据
models = ['原始模型', 'LoRA 微调后']
bleu_scores = [0.45, 0.60]
rouge_scores = [0.38, 0.55]
perplexities = [110, 85]

# 创建图表
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# 图1：BLEU分数
axs[0].bar(models, bleu_scores, color=['gray', '#BDEBDA'])
axs[0].set_title("BLEU 分数提升")
axs[0].set_ylabel("BLEU 分数")
axs[0].set_ylim(0, 1)
for i, v in enumerate(bleu_scores):
    axs[0].text(i, v + 0.02, f"{v:.2f}", ha='center')

# 图2：ROUGE-L分数
axs[1].bar(models, rouge_scores, color=['gray', '#D4BFFA'])
axs[1].set_title("ROUGE-L 分数提升")
axs[1].set_ylabel("ROUGE-L 分数")
axs[1].set_ylim(0, 1)
for i, v in enumerate(rouge_scores):
    axs[1].text(i, v + 0.02, f"{v:.2f}", ha='center')

# 图3：困惑度比较
axs[2].bar(models, perplexities, color=['gray', '#90CBC8'])
axs[2].set_title("困惑度对比（越低越好）")
axs[2].set_ylabel("困惑度")
axs[2].set_ylim(0, 120)
for i, v in enumerate(perplexities):
    axs[2].text(i, v + 2, f"{v}", ha='center')

plt.tight_layout()
plt.show()
```
